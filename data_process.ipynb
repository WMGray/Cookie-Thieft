```python
import os
import json
from collections import Counter
import re
def read_raw_file(file_path):
    return json.load(open(file_path, 'r'))
```

```python
def extract_group_all(file_folder):
    """
    Get statistics of entities types
    """
    res = []
    for file_path in os.listdir(file_folder):
        if not file_path.endswith('.json'):
            continue

        res.extend(_extract_group_one(os.path.join(file_folder, file_path)))

    return Counter(res)
```

```python
def _extract_group_one(file_path):
    data = read_raw_file(file_path)

    groups = []
    for i in range(len(data)):
        annotations = data[i]['annotations']
        annotations = annotations[len(annotations)-1] # select the newest annotations when multiple annotations exist

        for single_entity in annotations['result']:
            groups.extend(single_entity['value']['labels'])
            
    return groups
```

```python
import pandas as pd
import os
def _extract_text_one(file_path):
    data = read_raw_file(file_path)
    sample_list = []
    object_list = []
    group_list = []
    
    for i in range(len(data)):
        para = data[i]['data']['text'].replace("\t", ' ')
        annotations = data[i]['annotations']
        annotations = annotations[
            len(annotations) - 1]  # select the newest annotations when multiple annotations exist
        sample = data[i]['file_upload']
        sample = sample[sample.find("-")+1 : sample.find(".")]
        info_list = []
        for single_entity in annotations['result']:
#             print(single_entity)
            text = single_entity['value']['text']
            group = single_entity['value']['labels'][0]
            start_index = single_entity['value']['start']
#             print(text, group, start_index)
#             print("\n")
            info_list.append({'text': text, 'group': group, 'start_index': start_index})
        # 根据 start_index 进行排序
        info_list = sorted(info_list, key=lambda x: x['start_index'])
        
        # 提取信息
        text = ", ".join([x['text'] for x in info_list])
        group = ", ".join([x['group'] for x in info_list])
        
        sample_list.append(sample)
        object_list.append(text)
        group_list.append(group)
            
    return sample_list, object_list, group_list

def extract_text_all(file_folder):
    """
    Get statistics of entities types
    """
    for file_path in os.listdir(file_folder):
        if not file_path.endswith('.json'):
            continue
        sample_list , object_list, group_list= _extract_text_one(os.path.join(file_folder, file_path))
        print(len(object_list), len(group_list))
        data = pd.DataFrame({
            'sample': sample_list,
            'label': [1 if "abnormal" in file_path else 0] * len(sample_list),
            "object": object_list,
            'group': group_list
        })
        print(data.shape)
        data.to_csv(f"data/{file_path[:-5]}.csv", index=False)
    return 
data =  extract_text_all("data/final/")
```
